## Create and load sample dataframes
- run just once and then load

```python
# add 10 nodes
nodes_list = []
num_nodes = 100
for i in range(num_nodes):
    nodes_list.append({'node_id': i, 'twitter_user_id': i, 'twitter_username': 'abcd', 'bot_probability': np.random.random()})

# add 3*num_nodes edges randomly
num_edges = 3 * num_nodes
edges_list = []
for i in range(num_edges):
    edges_list.append({'edge_id': i, 'source': np.random.randint(0, num_nodes), 'target': np.random.randint(0, num_nodes), 'weight': 0.5})

nodes_df = pd.DataFrame(nodes_list)
edges_df = pd.DataFrame(edges_list)

# dump to csv
nodes_df.to_csv('./data/nodes.csv', index=False)
edges_df.to_csv('./data/edges.csv', index=False)
```

```python
# load from csv
nodes_df = pd.read_csv('./data/nodes.csv')
edges_df = pd.read_csv('./data/edges.csv')
```
